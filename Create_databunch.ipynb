```python
def open_arr_image(fn, **kwargs):
    """ Converts numpy array to fastai Image with 3 channels """
    # It has one channel, get that
    arr = np.load(fn)
    data = torch.from_numpy(arr)
    # Stack 3 copies as new channels
    data = torch.stack([data, data, data])
    return Image(data)

# Use function to read images
vision.data.open_image = open_arr_image
```

```python
cats = []
img_files = []

DIR_IMGS = Path('/home/rubn/Desktop/COVID')

# Save all spectrograms
for k, v in resp_dt[0].items():
  cats += [k] *len(v)
  for i, img in enumerate(v):
    fn = f'{k}_{i}.npy'
    print(f'Saving {fn}')
    img_files += [fn]
    np.save(DIR_IMGS/fn, img[0][:,:,0])

# We've saved all image files and corresponding categories
len(img_files), len(cats)
```

```python
# Make a dataframe out of filenames and cats
df1 = pandas.DataFrame(zip(img_files, cats))
df1.insert(2, 'is_test', False)
df1.head()
```

```python
cats = []
img_files = []

DIR_IMGS = Path('/home/rubn/Desktop/COVID')

# Save all spectrograms
for k, v in resp_dt[1].items():
  cats += [k] *len(v)
  for i, img in enumerate(v):
    fn = f'test_{k}_{i}.npy'
    print(f'Saving {fn}')
    img_files += [fn]
    np.save(DIR_IMGS/fn, img[0][:,:,0])

# We've saved all image files and corresponding categories
len(img_files), len(cats)
```

```python
df2 = pandas.DataFrame(zip(img_files, cats))
df2.insert(2, 'is_test', True)
df2.head()
```

```python
df = pd.concat([df1, df2])
```

```python
df.head
```

```python
df.to_pickle(DIR_IMGS/'COVID_resp_df.pkl')
df.head()
```

```python
# Now we can turn the df into a databunch
db = ImageList.from_df(df, path=DIR_IMGS).split_from_df().label_from_df().databunch().normalize(imagenet_stats)
    
# Serialize databunch
db.save(DIR_IMGS/'COVID_resp_db.pkl')
db.show_batch()
```
